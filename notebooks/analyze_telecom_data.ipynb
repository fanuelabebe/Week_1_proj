```python
df_clean['added_ul_dl_gaming'] = df_clean['Gaming UL (Bytes)'] +  df_clean['Gaming UL (Bytes)']
df_clean['added_ul_dl_social'] = df_clean['Social Media DL (Bytes)'] +  df_clean['Social Media UL (Bytes)']
df_clean['added_ul_dl_google'] = df_clean['Google DL (Bytes)'] +  df_clean['Google UL (Bytes)']
df_clean['added_ul_dl_email'] = df_clean['Email DL (Bytes)'] +  df_clean['Email UL (Bytes)']
df_clean['added_ul_dl_youtube'] = df_clean['Youtube DL (Bytes)'] +  df_clean['Youtube UL (Bytes)']
df_clean['added_ul_dl_netflix'] = df_clean['Netflix DL (Bytes)'] +  df_clean['Netflix UL (Bytes)']                                                               
df_clean['added_ul_dl_other'] = df_clean['Other DL (Bytes)'] +  df_clean['Other UL (Bytes)']
                                                               
# df_check = df[['Gaming UL (Bytes)','Gaming DL (Bytes)','added_ul_dl']]
# df_check
agg = df_clean.groupby('MSISDN/Number')[['added_ul_dl_gaming','added_ul_dl_social','added_ul_dl_google','added_ul_dl_email'\
                                  ,'added_ul_dl_other','added_ul_dl_netflix','added_ul_dl_youtube','Dur. (ms).1']].aggregate('sum')
agg_df = pd.DataFrame(agg)
agg_df['count'] = df['MSISDN/Number'].value_counts()
agg_df
```

```python
agg_df[['added_ul_dl_social','added_ul_dl_google']].sum().plot.bar(figsize=(15, 7.5))
```
